```python
import pandas as pd

# Load the dataset
file_path = "media/final_disease_symptoms_dataset.csv"  # Update with your actual file path
df = pd.read_csv(file_path)

# Ensure the "Disease" column exists and extract unique values
if "Disease" in df.columns:
    unique_diseases = df.drop_duplicates(subset=["Disease"])

    # Select 500 unique diseases or the maximum available
    sample_size = min(500, len(unique_diseases))
    df_sampled = unique_diseases.sample(n=sample_size, random_state=42)

    # Save the sampled dataset
    output_file = "sampled_500_diseases.csv"
    df_sampled.to_csv(output_file, index=False)
    print(f"CSV file with 500 unique diseases saved as {output_file}")
else:
    print("Error: 'Disease' column not found in the dataset.")
```
